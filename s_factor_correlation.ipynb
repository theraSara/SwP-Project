```python
import pandas as pd
import numpy as np

rt_df = pd.read_csv("data/SPRT_LogLin_216.csv")
diag_df = pd.read_csv("data_extraction/bk21_stimuli_with_cloze.csv")
```

```python
rt_df['condition'] = rt_df['condition'].str.upper()
diag_df['condition'] = diag_df['condition'].str.upper()
```

```python
merged = pd.merge(rt_df, diag_df, on=['ITEM', 'condition'])
merged['log_SUM_3RT'] = np.log(merged['SUM_3RT'])
```

```python
s_factors = [50, 100, 200, 500, 1000, 2000]
results = []
for s in s_factors:
    col = f'cloze_surprisal_s{s}'
    col_sq = f'{col}_sq'
    
    # correlation (Pearson r)
    corr_linear = merged['log_SUM_3RT'].corr(merged[col])
    corr_squared = merged['log_SUM_3RT'].corr(merged[col_sq])
    
    results.append({
        'S_Factor': s,
        'Linear_Corr': corr_linear,
        'Squared_Corr': corr_squared
    })

res_df = pd.DataFrame(results)
print("Correlation Results (Target: log(SUM_3RT)):")
print(res_df.sort_values(by='Squared_Corr', ascending=False))
```

From this data: *linear surprisal* with a higher smooting factor better captured the variance in this specific dataset (BK21). However, for cross-study consistency, will proceed with the **S=200 smoothing factor** as the primary human baseline.

```python
winner_idx = res_df['Squared_Corr'].idxmax()
winner = res_df.loc[winner_idx]

print(f"\nWINNER: S={int(winner['S_Factor'])} (Squared)")
print(f"Correlation: r = {winner['Squared_Corr']:.4f}")
```
